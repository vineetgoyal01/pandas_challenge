```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
Purchase_data = purchase_data_df["SN"].unique()
```

## Player Count

* Display the total number of players


```python
Total_Players = len(Purchase_data)
print(f"Total_Players : {Total_Players}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
total_items = purchase_data_df["Item ID"]
count_total_items = len(total_items)
item_list = purchase_data_df["Item ID"].unique()
number_of_unique_items = len(item_list)
total_price = purchase_data_df["Price"].sum()
average_price = (total_price/count_total_items).round(2)
total_purchase = len(purchase_data_df["Price"])
print(f" Number of Unique Items : {number_of_unique_items} {total_purchase}")
```

```python
# Creating a summary DataFrame using above values
summary_df = pd.DataFrame({"Unique Items" :[number_of_unique_items],
                           "Average Price" :[average_price],
                           "Total Number of Purchases" : [total_purchase],
                           "Total Amount" : [total_price]  
                           
                           
    
    
})
summary_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age



* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below



* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_df["Age Demo"]= pd.cut(purchase_data_df["Age"], age_bins, labels=group_names)

age_analysis_df = purchase_data_df[['Price', 'Age Demo']].groupby(['Age Demo']).count()
age_analysis_df['Average Purchase Price'] = purchase_data_df[['Price', 'Age Demo']].groupby(['Age Demo']).mean()
age_analysis_df['Total Purchase Price'] = purchase_data_df[['Price', 'Age Demo']].groupby(['Age Demo']).sum()

age_analysis_df['Average Purchase Price'] = age_analysis_df['Average Purchase Price'].round(2)
age_analysis_df['Total Purchase Price'] = age_analysis_df['Total Purchase Price'].round(2)

age_analysis_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_items = purchase_data_df.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items.columns = ['Purchase Count', 'Total Purchase Value']

most_items.reset_index(inplace=True)
most_items.set_index(['Item ID','Item Name'] ,inplace=True)
most_items = most_items[['Purchase Count', 'Price', 'Total Purchase Value']]


most_popular_items = most_items.sort_values(by='Purchase Count', ascending=False)


most_popular_items.head()
```

## Most Profitable Items
